## Area plot 
berwarna transparansi

```python
df_top5 = df_canada.sort_values('Total', ascending= False)
df_top5 = df_top5.head(5)
df_top5 = df_top5[years].T

df_top5.plot(kind="area", stacked=False,alpha = 0.35, figsize= (15,7))
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Top 5 Countries Trend")
plt.xticks(years[::2])
plt.show()
```

```python
# stacked nya dihilangin
df_top5 = df_canada.sort_values('Total', ascending= False)
df_top5 = df_top5.head(5)
df_top5 = df_top5[years].T

df_top5.plot(kind="area",alpha =.8, figsize= (15,7))
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Top 5 Countries Trend")
plt.xticks(years[::2])
plt.show()
```

```python
df_bottom5 = df_canada.sort_values('Total', ascending= False)
df_bottom5 = df_bottom5.tail(5)
df_bottom5= df_bottom5[years].T

df_bottom5.plot(kind="area",alpha =.45, figsize= (10,8))
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Top 5 Countries Contribute The Least Immigrants")
plt.xticks(years[::2], rotation =90)
plt.show()
```

## Anotate

```python
iceland = df_canada.loc["Iceland", years]
iceland.plot(kind="bar", figsize=(10,5))
plt.title("Immigrants from Iceland")
plt.xlabel("Year")
plt.ylabel("Num of Immigrants")

plt.annotate("",xy=(32,70), xytext=(26.5,20), xycoords="data",
            arrowprops={"arrowstyle":"-|>", "connectionstyle": "arc3", "color": "blue", "lw":2}
            )

plt.annotate(
    "2008-2011 Financial Crisis", 
    xy=(26, 25), 
    rotation=62,
    va="bottom",
    ha="left",
)

plt.show()
```

```python
df_top15 = df_canada.sort_values("Total", ascending = False)["Total"].head(15)
df_top15.plot(kind="barh", figsize=(8,10))
plt.title("Top 15 Countries Contributing to Imigration to Canada")
plt.xlabel("Number of Immigrants")
plt.show()
```

## pie charts
merepresentasikan suatu kategori atau suatu negara

```python
df_continents= df_canada.groupby("Area").sum()
df_continents["Total"].plot(kind="pie", 
                            figsize=(15,6), autopct="%1.1f%%",
                            labels=None, 
                            pctdistance= 1.2,
                            explode=[0.1,0,0,0,0.1,0.1])
plt.axis("equal")
plt.title("Immigration by Area (1980 -2013)")
plt.legend(labels=df_continents.index, loc = "upper left")
plt.show()
```

## Subplot

```python
# sediain figure dengan canvas 1 baris 2 kolom
fig = plt.figure(figsize=(12, 6))
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

china_india.T.plot(kind="box", vert=False, ax=ax0)
ax0.set_title("Box plots of Immigrants from China and India", fontdict={"fontsize":12})
ax0.set_xlabel("Num of Immigrant")
ax0.set_ylabel("Countries")

china_india.T.plot(kind="line", ax=ax1)
ax1.set_title("Immigrants Growth of China and India to Canada", fontdict={"fontsize":12})
ax1.set_xlabel("Years")
ax1.set_ylabel("Num of Immigrants")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(211) # 2 baris 1kol 1 kol
plt.bar(iceland.index, iceland.values)
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Canada")

plt.subplot(212)
plt.barh(df_top15.index, df_top15.values)
plt.xlabel("Total Immigrants")
plt.ylabel("Countries")
plt.title("Top 15 Countries")

plt.show()
```
